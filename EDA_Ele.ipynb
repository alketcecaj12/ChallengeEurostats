```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data/nrg_cb_em_linear.csv.gz")
```

```python
data
```

```python
gbc = data.groupby("geo")
```

```python
for i, k in gbc: 
    print(i, len(k))
```

```python
at = gbc.get_group("AT")
at
```

```python
plt.hist(at.OBS_VALUE)
plt.show()
```

```python
at.nrg_bal.value_counts()
```

```python
gb_nrgbal = at.groupby("nrg_bal")
```

```python
for i, k in gb_nrgbal: 
    print(i)
    timeseries = k[["TIME_PERIOD","OBS_VALUE"]].set_index("TIME_PERIOD")[:75]
    plt.figure(figsize = (20, 4))
    plt.plot(timeseries)
    plt.xticks(rotation = 45)
    plt.show()
```
